```python
arr = df_accounts.investor_type.values
arr2 = set()
for val in arr:
    val = str(val)
    tp = val.split("/")
    for val2 in tp:
        tp2 = val2.split("\\")
        for val3 in tp2:
            x = val3.replace(" ","")
            x = x.lower()
            arr2.add(x)

for col in arr2:
    df_accounts[col] = 0

for i,val in enumerate(df_accounts.investor_type):
    val = df_accounts.iloc[i]["investor_type"]
    val = str(val)
    tp = val.split("/")
    for val2 in tp:
        tp2 = val2.split("\\")
        for val3 in tp2:
            x = val3.replace(" ","")
            x = x.lower()
            df_accounts.set_value(i,x,1)

df_accounts.drop('investor_type',axis=1,inplace=True)

for col in df_accounts.columns:
    if df_accounts[col].dtype.name == "bool":
        df_accounts[col] = pd.Categorical(df_accounts[col])
        df_accounts[col] = df_accounts[col].cat.codes

df_accounts.to_csv('Accounts_1.csv',index=False)
```

```python
def parseBuildingTaxExpenses(expense,i):
    new_exp = ""
    expense = str(expense)
    expense = expense.strip()
    expense = expense.replace(',',';')
    expense = expense.replace('$','')
    x = expense.split(';')
    data_dict = {}
    if x[0] != "nan":
        for val in x:
            y = val.split('@')
            y1 = y[0].strip().split(' ')
            y2 = y[1].strip().split('/')
            year = int(y1[0].strip())
            if y2[-1].strip().lower() == "unit" or y2[-1].strip().lower() == "sf":
                unit = y2[-1].strip()
            else:
                unit = ""
            price = float(y2[0].strip())
            y1 = y1[1:]
            tax = " ".join(y1)
            data_dict[tax] = [price,year,unit]
        x_unit = "unit"
        for key in data_dict.keys():
            if data_dict[key][2] != "":
                x_unit = data_dict[key][2]
        for key in data_dict.keys():
            data_dict[key][2] = x_unit
        return data_dict
    else:
        return None
```

```python
for i,row in enumerate(df_train.iterrows()):
    data_dict = parseBuildingTaxExpenses(row[1].building_tax_expenses,i)
    if data_dict != None:
        for key in data_dict.keys():
            value = data_dict[key]
            df_train.at[i,key] = value[0]
            df_train.at[i,key+"_year"] = value[1]
            df_train.at[i,value[2]] = 1
```
